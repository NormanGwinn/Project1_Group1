```python
#--Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
#--Point to CSV file
path = "Resources/Confirmed_Case.csv"
world_path = "Resources/world_pop_csv.csv"
```

```python
#--Read CSV file
confirmed_df = pd.read_csv(path)
world_poo = pd.read_csv(world_path)
```

```python
# confirmed_df = confirmed_df.reset_index(drop = True)
transpose_df = confirmed_df.T
transpose_df.index = [pd.Timestamp(d) for d in transpose_df.index]
# Now sort the DataFrame by the Timestamp index
transpose_df = transpose_df.loc[transpose_df.index.sort_values(),:]
npNaN = np.empty(transpose_df.shape)
npNaN[:] = np.NaN
rel_country = pd.DataFrame(npNaN, columns=transpose_df.columns.values)


# # Shift the trends up, to start on relative day zero
for (the_column, the_data) in transpose_df.iteritems():
#     print(type(the_column), type(the_data),the_data)
    trend = np.trim_zeros(the_data.to_numpy(),'f')
    rel_country[the_column][0:len(trend)] = trend
# #     break
    
# transpose_df
rel_country
```

```python
# GroupBy Country and sum(number of cases)
group_country = pd.DataFrame(confirmed_df.groupby("country", as_index = False).sum())
# group_country.head(5)
group_country.shape
```

```python
# Calculate (column i+i) - (column i)
minus_confirm_df = group_country.diff(axis = 1)
minus_confirm_df = minus_confirm_df.drop(columns = ['1/22/2020','country'], axis = 1)
minus_confirm_df.head(30)
```
